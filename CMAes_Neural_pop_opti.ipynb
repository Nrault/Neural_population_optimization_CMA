```python
'''
Function CreateRampAndHoldStim 
    Create a stimulus composed of a constant direction and a ramp and hold amplitude of whisker deflection with a 
    timestep of 1-ms
    
    Input :
        -startTime : The time at which the ON ramp start (beginning of the increase in amplitude)
        -endTime   : The time at which the OFF ramp ends (ending of the decreases in amplitude)
        -maxamp    : The value of amplitude at which the ramp shall culminate and be maintained during the 
                     hold phase
        -minamp    : The minimal amplitude of the stimulus (most likely 0, which means resting)
        -angle     : The direction of the whisker deflection
        -rampTime  : The duration of the ramp phases
        -duration  : The duration of the entire simtulus
        
    Output :
        -direction : A timed array containing the angle of the stimulus for each timestep
        -amp       : A timed array containing the amplitude of the stimulus for each timestep
'''
def CreateRampAndHoldStim(startTime, endTime, maxamp, minamp, angle, rampTime, duration):
    #Ramp and hold paradigm, we assume that the direction is constant therefore at timestep 0
    #direction is equal to the angle of the stimulus
    direction = [angle]
    #Whisker starts at resting position in ramp and hold
    amp = [minamp]
    #For each timestep in the simulation
    for i in range(1, duration):
        #The direction is constant, therefore we add it to the array
        direction.append(angle)
        #If the current timestep is comprised between the begining of the starting time of the ramp
        #and the ending time of the ramp (ending time - time required to go back to resting position)
        if i >= startTime and i < (endTime - rampTime):
            #If the previous amplitude is lower than the maximum amplitude
            if amp[i-1] < maxamp :
                #We increase the amplitude
                amp.append(amp[i-1] + maxamp/rampTime)
            #Else we stay at maximum amplitude
            else :
                amp.append(maxamp)
        else:
            #If we are at the end of the hold stimulus and the amplitude is above the resting state (0)
            if i >= (endTime - rampTime) and amp[i-1] > minamp:
                #We decrease the amplitude
                amp.append(amp[i-1] - maxamp/rampTime)
            #Otherwise we add the minimu amplitude to the array
            else:
                amp.append(minamp)
    #We convert direction and amp into timed array format so that it can be used in BRIAN neuron equations
    direction = TimedArray(direction, dt=1*ms)
    amp = TimedArray(amp, dt=1*ms)
    return direction, amp
```

```python
'''
Function CreateRampAndHoldStim 
    Create a stimulus composed of a constant direction and a ramp and hold amplitude of whisker deflection with a 
    timestep of 1-ms
    
    Input :
        -startTime : The time at which the ON ramp start (beginning of the increase in amplitude)
        -endTime   : The time at which the OFF ramp ends (ending of the decreases in amplitude)
        -maxamp    : The value of amplitude at which the ramp shall culminate and be maintained during the 
                     hold phase
        -minamp    : The minimal amplitude of the stimulus (most likely 0, which means resting)
        -angle     : The direction of the whisker deflection
        -rampTime  : The duration of the ramp phases
        -duration  : The duration of the entire simtulus
        
    Output :
        -direction : A timed array containing the angle of the stimulus for each timestep
        -amp       : A timed array containing the amplitude of the stimulus for each timestep
        -amp_on    : A timed array containing 1 if the amplitude is above 0 or 0 otherwise for each timestep
'''
def CreateRampAndHoldStimAmpOnOff(startTime, endTime, maxamp, minamp, angle, rampTime, duration):
    #Ramp and hold paradigm, we assume that the direction is constant therefore at timestep 0
    #direction is equal to the angle of the stimulus
    direction = [angle]
    #Whisker starts at resting position in ramp and hold
    amp = [minamp]
    #No amplitude applied at the begining of the simulation
    amp_on = [0]
    for i in range(1, duration):
        #If we are in the ramp and hold phase then there is an applied amplitude (1)
        if i >= startTime and i < endTime:
            amp_on.append(1)
        #Otherwise there is no applied amplitude (0)
        else:
            amp_on.append(0)
        #The direction is constant, therefore we add it to the array
        direction.append(angle)
        #If the current timestep is comprised between the begining of the starting time of the ramp
        #and the ending time of the ramp (ending time - time required to go back to resting position)
        if i >= startTime and i < (endTime - rampTime):
            #If the previous amplitude is lower than the maximum amplitude
            if amp[i-1] < maxamp :
                #We increase the amplitude
                amp.append(amp[i-1] + maxamp/rampTime)
            #Else we stay at maximum amplitude
            else :
                amp.append(maxamp)
        else:
            #If we are at the end of the hold stimulus and the amplitude is above the resting state (0)
            if i >= (endTime - rampTime) and amp[i-1] > minamp:
                #We decrease the amplitude
                amp.append(amp[i-1] - maxamp/rampTime)
            #Otherwise we add the minimu amplitude to the array
            else:
                amp.append(minamp)
    #We convert direction, amp and amp_on into timed array format so that it can be used in BRIAN neuron equations
    direction = TimedArray(direction, dt=1*ms)
    amp_on = TimedArray(amp_on, dt=1*ms)
    amp = TimedArray(amp, dt=1*ms)
    return direction, amp, amp_on
```

```python
'''
Function DefineFavAngle_nat_oriented
    Create a uniformly distributed array of angles oriented towards naturally plausible stimuli for the 
    whisker system of the rodent (forward, backward as it is unlikely that most of the stimuli comes from
    above or under) around which the neurons will be spread. 50% will be disrtibuted forward, 50% will be 
    distributed backward 
    
    Input :
        forward_low   : the lowest possible angle for stimuli coming forward
        forward_high  : the highest possible angle for stimuli coming forward
        backward_low  : the lowest possible angle for stimuli coming backward
        backward_high : the highest possible angle for stimuli coming backward
        nb_neuron     : the amount of angles to determine (1 per neuron)
    Return :
        fav_angle     : An array that contains the angles of stimulus that elicitate the most activity
                        for each mechanoreceptors
'''
def DefineFavAngle_nat_oriented(forward_low, forward_high, backward_low, backward_high, nb_neuron):
    #For each neuron in half of the population we pick a random integer between the lowest forward angle
    #and the highest forward angle
    #Example : take a circle with 0 degree on the top, forward low = 45 anf forward high = 135,
    #          half of the population will have a favorite angle between 45 and 135 degrees
    forward = [rng.randrange(forward_low, forward_high, 1) for i in range(0, int(nb_neuron/2))]
    #For each neuron in the other half of the population we pick a random integer between the 
    #lowest backward angle and the highest backward angle
    #Example : take a circle with 0 degree on the top, backward low = 225 anf forward high = 315,
    #          half of the population will have a favorite angle between 225 and 315 degrees
    backward = [rng.randrange(backward_low, backward_high, 1) for i in range(0, int(nb_neuron/2))]
    #We combine both of the previously obtained array into one
    fav_angle = np.concatenate((np.asarray(forward), np.asarray(backward)))
    return fav_angle
```

```python
#adexp model of mechanoreceptors in the whisker follicle of the mouse
#if we represent the follicle as a cylinder, the fav_direction is the angle of the center of the 
#mechanoreceptor on this cylinder, direction is the direction of the applied force to the whisker
#and amplitudeOnOff is wether or not a force is applied to the whisker
#equation and name of variable from An Introductory Course in Computational Neuroscience by Paul Miller 2018

#Iapp is the applied current I filtered by the receptive field
#ISRA_SA1 is the adaptation
#Il is the leak current
eq_adexp_SA1_tg = '''
Iapp = I * (amplitudeOnOff(t)) * ((1 + cos((direction(t) - fav_direction) * (pi/180)))/2) : amp
dISRA_SA1/dt = (a_SA1 * (v  - El_tg) - ISRA_SA1)/tau_SRA1 : amp
Il = Gl_tg * (El_tg - v + delta_th * exp((v - Vth_tg)/delta_th)) : amp
dv/dt = (Il - ISRA_SA1 + Iapp)/Cm : volt

fav_direction : 1
I : amp
'''
```

```python
'''
Function MSE_model_relative
    Compute the mean squared error between the model's normalized results and the normalized data
    
    Input :
        data  : the data to fit
        model : the normalized resutls from the model
        
    Output :
        MSE   : the mean squared error
'''
def MSE_model_relative(data, model):
    idx = []
    #For each datapoint
    for t in data[0] :
        #We retrieve the corresponfing timestep in our model 
        idx.append(int(t))
    #We retrieve our model's value at each timestep corresponfing to a datapoint
    model = model[idx]
    #We comput the mean squared error
    MSE = np.square(np.subtract((data[2]/max(data[2])), model)).mean()
    return MSE
```

```python
'''
Function MSE_model_relative
    Compute the mean squared error between the model's results and the data
    
    Input :
        data  : the data to fit
        model : the resutls from the model
        
    Output :
        MSE   : the mean squared error
'''
def MSE_model(data, model):
    idx = []
    #For each datapoint
    for t in data[0] :
        #We retrieve the corresponfing timestep in our model 
        idx.append(int(t))
    #We retrieve our model's value at each timestep corresponfing to a datapoint
    model = model[idx]
    #We comput the mean squared error
    MSE = np.square(np.subtract(data[2], model).mean())
    return MSE
```

```python
'''
Function model_pop_SA1
    Initialize and run a model of neural population with BRIAN
    
    Input :
        a_SA1, b_SA1, tau_SRA1 : variables relative to the adexp model of neuron
        direction              : a timed array which contains the direction of the stimulus for
                                 each timestep
        amplitudeOnOff         : a timed array which contains 1 if an amplitude is applied, 0
                                 otherwise for each timestep
        nb_input_neuron        : size of the input layer (here the only layer)
        iteration              : the number of the current iteration
        pop_size               : the number of trial per pool of optimization
        data                   : the data to fit
        fav_direction          : list of favorite angle for each input neuron (size = nb_input_neuron)
'''

def model_pop_SA1(a_SA1, b_SA1, tau_SRA1, direction, amplitudeOnOff, nb_input_neuron, iteration, pop_size, data, fav_direction):
    #start_scope ensure that the model run in its own environment and doesn't use variables from previously ran
    #models
    start_scope()

    #Initialization of our mechanoreceptor model
    El_tg = -50 * mV
    Gl_tg = 10 * nS
    Cm = 0.1 * nF
    Vth_tg = -30 * mV
    Vreset_tg = -45 * mV
    delta_th = 5 * mV

    #Creation of our population
    mecha_SA1 = NeuronGroup(nb_input_neuron, eq_adexp_SA1_tg, method = 'euler', threshold = 'v > Vth_tg', 
                         reset = '''v = Vreset_tg
                                    ISRA_SA1 = ISRA_SA1 + b_SA1''')
    
    statemon = StateMonitor(mecha_SA1, variables=True, record=True)
    rt = PopulationRateMonitor(mecha_SA1)

    #Initialization of our neuron population
    mecha_SA1.v = El_tg
    mecha_SA1.I = 900 * pA
    mecha_SA1.fav_direction = fav_direction

    #Simulation
    run(4*second)
    
    #Ploting of the result if this is the last iteration of the pool
    if iteration ==  pop_size-1:
        plot(statemon.t/ms, rt.rate)
        plot(data[0], data[2])
        #plot(statemon.t/ms + 810, reversed_rate/ max(rt.rate) * 0.45)
        xlim(0, 4000)
        xlabel('Time (ms)')
        ylabel('nb spike');
        show()

    return rt.rate/max(rt.rate)
    #return rt.smooth_rate(width = 10 * ms)/max(rt.smooth_rate(width = 10 * ms))
```

```python
#initialization of the CMA
#mean : the mean value for each variable to optimize
#sigma : the variation
#bounds : the boundaries ([down, up]) for each variable to optimize
optimizer = CMA(mean=np.array([50.0, 100.0, 200.0]), sigma=50.0, bounds=np.array([[0.0, 100.0], [0.0, 200.0], [0.0, 2000.0]]))
#Creation of the stimulus, here we retrieve the direction and the on/off array of amplitude (see model of neuron)
direction, _, amplitude = CreateRampAndHoldStimAmpOnOff(min(Data_SA1_raw[0]), max(Data_SA1_raw[0]), 90, 0, 90, 100, 4000)
#We assume here that half of the population respond to a deflection at max, we therefore consider the max response
#of the data to be half of the total population, therefore total population = max data * 2
nb_input_neuron = int(max(Data_SA1_raw[2]) * 2)
#We create an array of favorite directio :
#50% of the population has a favorite angle between 45 and 135
#50% of the population has a favorite angle between 225 and 315
fav_direction = DefineFavAngle_nat_oriented(45, 135, 225, 315, nb_input_neuron)
#We run the optimizer for 50 generations
for generation in range(50):
    solutions = []
    #for each generation we run population_size model
    for it in range(optimizer.population_size):
        #We ask the optimizer to assign values for each of our parameter defined in the first line of this cell
        x = optimizer.ask()
        #We run the model
        model = model_pop_SA1(x[0] * nS, x[1] * pA, x[2] * ms, direction, amplitude, nb_input_neuron-1, 
                              it, optimizer.population_size, Data_SA1_raw, fav_direction)
        #We compute the MSE
        value = MSE_model_relative(Data_SA1_raw, model)
        #We add the the parameters and the MSE to the pull of solutions
        solutions.append((x, value))
        print(f"#{generation} {value} (a = {x[0]}, b = {x[1]}, tau = {x[2]})")

    print('---------------------------------------------------------------------')
    #We updated the optimizer's strategy
    optimizer.tell(solutions)
```
